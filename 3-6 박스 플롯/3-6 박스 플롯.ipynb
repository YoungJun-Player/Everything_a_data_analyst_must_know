상자 수염 그림이라고도 불리는 박스 플롯은 네모 상자 모양에 최댓값과 최솟 값을 나태는 선이 결합된 모양의 데이터 시각화 방법이다

박스 플롯은 하나의 그림으로 양적 척도 데이터의 분포 및 편향성, 평균값과 중앙값 등 다양한 수치를 보기 쉽게 정리해준다

최솟값 : 제1사분위에서 1.5 IQR을 뺀 위치

제1사분위(Q1) : 25%의 위치

제2사분위(Q2) : 50%의 위치 (중앙값(median)을 의미)

제3사분위(Q3) : 75%의 위치

최댓값: 제3사분위에서 1.5 IQR을 더한 위치

박스 플롯을 해석할 때는 항상 데이터 분포도를 함께 떠올리는 습관이 필요하다

박스플롯 시각화

1. 패키지 임포트

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

2. 데이터 불러오기 및 확인

```python
df= pd.read_csv('/content/50_Startups.csv')

df.head()
```

3. 기본 박스플롯 시각화

```python
# 세로 박스 플롯

plt.figure(figsize=(8,6))
sns.boxplot(y='Profit', data =df)
plt.show()

# 가로 박스 플롯

plt.figure(figsize=(8,2))
sns.boxplot(x='Profit', data = df)
plt.show()
```

4. 그룹별 박스 플롯 시각화

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='State', y='Profit', data=df)
plt.show()
```

5. 박스 플롯 추가 옵션 적용 (실제 데이터 포인트 추가 적용)

```python
sns.boxplot(x='State', y='Profit',
            showmeans = True,
            boxprops={'facecolor':'None'},
            data = df)

sns.stripplot(x='State', y='Profit',
              data = df,
              jitter = True,
              marker = 'o',
              alpha=0.5,
              color='black')

plt.show()
```
