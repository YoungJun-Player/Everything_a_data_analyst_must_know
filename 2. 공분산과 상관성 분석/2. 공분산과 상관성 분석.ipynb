1. 필요한 패키지 임포트

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

2. 데이터 불러오기 및 확인

```python
df = pd.read_csv('/content/wine-quality.csv')
```

3. 데이터 샘플 확인

```python
df.head()
```

4. 산점도 행렬 시각화

```python
sns.set(font_scale = 1.1)   #폰트 크기 설정
sns.set_style('ticks')  #축 눈금 설정
sns.pairplot(df,
             diag_kind='kde' #상관계수가 1이면 분포로 표시
             )

plt.show()
```

5. 공분산 확인

```python
# 문자열이 아닌 열만 선택
df_num = df.select_dtypes(include=[np.number])

# 선택된 숫자형 열로 공분산 계산
df_num.cov()
```

6. 피어슨 상관계수 산출

```python
# 문자열이 아닌 열만 선택
df_num = df.select_dtypes(include=[np.number])

# 선택된 숫자형 열로 피어슨 상관계수 계산
df_num.corr()
```

7. 상관계수 히트맵 시각화

```python
sns.heatmap(df_num.corr(), cmap='viridis')
```

8. 클러스터맵 히트맵 시각화

```python
sns.clustermap(df_num.corr(),
               annot = True,
               cmap = 'RdYlBu_r',
               vmin = -1, vmax = 1,
               )
```

total sulfur dioxide와 free sulfur dioxide 변수의 상관관계는 0.72로 강한 양의 상관관계

density와 alcohol은 -0.69로 강한 음의 상관관계 (와인의 밀도가 높을수록 알코올 도수가 낮음)

9. 클러스터맵 히트맵 중복영역 제거

```python
np.triu(np.ones_like(df_num.corr()))

mask = np.triu(np.ones_like(df_num.corr()))

fig, ax = plt.subplots(figsize=(15,10))

sns.heatmap(df_num.corr(),
            mask = mask,
            vmin = -1,
            vmax = 1,
            annot = True,
            cmap='RdYlBu_r',
            cbar = True)

ax.set_title('wine Quality Correlation', pad =15)
```
