산점도는 점들의 분포와 추세를 통해 두 변수 간의 관계를 파악할 수 있다


산점도를 그릴 때는 극단치를 제거하고서 그리는 것이 좋다

극단치로 인해 주요분포구간이 압축되어 시각화의 효율이 떨어지기 때문이다

데이터가 너무 많아서 점들이 서로 겹쳐서 정보를 제대로 확인하기 어려울 때 각각의 점에 투명도를 주어

점들의 밀도를 함께 표현할 수 있도록 할 수 있다.


또는 값의 구간을 나누어 빈도에 따른 농도나 색상을 다르게 하여 표현할 수도 있다


산점도는 두 개의 변수 간 관계만 표현할 수 있다는 단점이 있다

버블차트를 이용하면 세가지 요소의 상관관계를 표현할 수 있다

버블의 크기를 통해 한 가지 요소를 추가적으로 볼 수 있는 것이다

버블 차트는 원의 면적을 함께 봐야 하기 때문에 관측치가 너무 많아지게 되면 정보 전달의 효율이 떨어진다

관측치가 100개 이상 넘어갈 경우, 데이터를 축약하거나 다른 시각화 방법을 사용하는 것이 좋다

버블차트를 해석할 때는 원의 지름이 아닌 면적을 통해 크기를 판단하도록 주의해야 한다

지름이 두배가 크면 실제 크기는 네 배가 큰 것이기 때문이다

1. 패키지 임포트

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

2. 데이터 불러오기 및 확인

```python
df = pd.read_csv('/content/50_Startups.csv')

df.head()
```

3. 기본 산점도 시각화

```python
plt.scatter(df['R&D Spend'], df['Profit'],s=50, alpha=0.4)
plt.show()
```

s는 점크기, alpha는 투명도

R&D 비용이 증가할수록 수익도 선형적으로 증가하는 것을 확인할 수 있다

4. 산점도 회귀선 시각화

```python
ax = sns.lmplot(x='R&D Spend', y='Profit', data=df)
plt.show()
```

5. 버블차트 시각화

```python
plt.scatter(df['R&D Spend'], df['Profit'], s=df['Marketing Spend']*0.001,
            c=df['Administration'], alpha =0.5, cmap='Spectral')

plt.colorbar()
plt.show()
```

마케팅 비용은 버블의 크기로 표현되도록 설정했다


관리비용은 색상으로 표현되도록 설정하여 붉은 색상은 적은 비용, 푸른 색상은 많은 비용을 의미한다
