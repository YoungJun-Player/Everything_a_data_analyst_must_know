# 1. 패키지 임포트

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
import seaborn as sns
```

# 2. 데이터 불러오기 및 확인

```python
df = pd.read_csv('/content/glass.csv')

df.head()
```

# 3. 데이터 정규화

```python
# 차원 축소할 칼럼만 선택 (종속변수 Type 제거)
df1 = df.drop('Type', axis=1)

# 데이터 정규화 적용
MinMaxScaler = MinMaxScaler()
df_minmax = MinMaxScaler.fit_transform(df1)

# 칼럼명 결합
df_minmax = pd.DataFrame(data = df_minmax, columns = df1.columns)

df_minmax.head()
```

표준화 등 다른 스케일링 방법을 적용해도 된다

정규화를 적용하는 이유는 변수들 간의 스케일 차이가 클 경우 분산이 큰 변수가 주성분에 과도하게 영향을 많이 주게 되기 때문이다

# 4. 주성분 생성 후 설명력 확인

```python
# 주성분 개수 설정(최대 수 설정)
pca = PCA(n_components = 9)
df_pca = pca.fit_transform(df_minmax)

# 주성분으로 변형된 테이블 생성
df_pca = pd.DataFrame(data=df_pca, columns = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9'])

# 주성분 설명력 확인
np.round_(pca.explained_variance_ratio_, 3)
```

제1주성분은 45.4%, 제2주성분은 18%, 제3주성분은 12.6%의 설명력을 가
지고 있는 것을 확인할 수 있다. 9개 주성분의 설명력 총 합은 100%이다.


# 5. 2개의 주성분 생성

```python
# 주성분 설정 (2개)
pca = PCA(n_components = 2)
df_pca = pca.fit_transform(df_minmax)

# 주성분으로 변형된 테이블 생성
df_pca = pd.DataFrame(data=df_pca, columns = ['C1', 'C2'])

df_pca.head()
```

약 63%의 설명력을 가지는 두 개의 주성분까지 생성하여 테이블을 생성한다.

이로써 기존의 9개의 변수
를 2개의 변수로 차원축소를 했다

# 6. 주성분에 따른 종속변수 시각화

```python
# 주성분 테이블에 종속변수 칼럼 결합
df_concat = pd.concat([df_pca, df[['Type']]], axis=1)

# 산점도 시각화
sns.scatterplot(data = df_concat, x='C1', y='C2', hue='Type')
```

제 1주성분과 제2주성분을x축과y축으로 하여 종속변수인 Type이 어떻게 분포하는지 확인한다.
