관측치 A를 기준으로, B와 C중 어느 관측치가 더 가까이 있는가를 판단하기 위한 것이 데이터 측정이다

데이터 간의 거리측정 혹은 유사도 측정은 데이터 분석 및 모델링에서 매우 중요하다

분류모델이나 군집모델의 경우 거의 필수적으로 데이터 거리를 활용한다

추천 시스템의 경우에도 사용자나 아이템 간의 거리를 측정하여

유사성이 높은 사용자나 아이템을 추천해 준다

데이터 거리를 측정하기 전에 데이터 표준화나 정규화 가공을 해줘야 한다

# 유클리드 거리

피타고라스 정리를 활용한 것

즉 관측치 간의 직선거리를 측정하는 것

값이 0에 가까울수록 데이터 간 거리가 짧다는 것이므로 유사도가 높다는 것을 의미

# 맨해튼 거리

맨해튼의 격자모양 도로에서 최단 거리를 구하는 원리를 이용한다

A지점과 B지점까지의 X축거리, Y축거리를 합해주기만 하면 된다.

# 민코프스키 거리

옵션값을 설정하여 거리 기준을 조정할 수 있는 거리 측정 방법이다

유클리드 거리 수식과 동일하며 단지 제곱 부분을 p-norm 값으로 조정할 수 있다

p값을 1로 설정하면 맨해튼 거리와 동일하고, 2로 설정하면 유클리드 거리와 동일하다

p값은 반드시 1 이상이어야 하고 정수가 아니어도 상관없다

# 체비쇼프 거리

민코프스키 거리의 p값을 무한대로 설정했을 때

체비쇼프 거리 혹은 맥시멈거리라고 한다

체비쇼프 거리는 군집 간의 최대 거리를 구할 때 사용한다

일반적으로 잘 사용되지는 않지만 때에 따라서 군집 간 최대 거리가 중요한 경우에 사용할 수 있다

이 역시 계산 값이 0에 가까울 수록 유사한 것이다

# 마할라노비스 거리

마할라노비스 거리는 유클리드 거리에 공분산을 고려한 거리 측정 방법이다

변수 내 분산과 변수 간 공분산을 모두 반영하여 A와 B간 거리를 계산한다

따라서 단순 거리에 상관성을 함께 볼 수 있다는 장점이 있다

마할라노비스 거리는 X와 Y의 공분산을 고려하여 거리를 측정한다

마할라노비스 거리는 확률 분포를 고려하기 때문에 공분산 행렬을 사용한다

공식은 선형 대수 개념이 필요하기 때문에 이해하기가 까다롭다



# 코사인 거리

코사인 유사도를 먼저 이해해야 한다

코사인 유사도는 벡터 사이의 각도만으로 두 점 간의 유사도를 측정한다

즉 두 벡터의 사이각을 구해서 유사도를 구하는 것이다

두 점 간의 각도가 작으면 유사도가 높고, 각도가 크면 유사도가 낮아진다

실제 거리보다, 좌표 공간상 각도가 얼마나 차이나는가에 따라 유사도를 측정한다

코사인 유사도는 -1에서 1사이의 값을 가지며, 두 벡터의 방향이 완전이 동일하면 1의 값을 가진다

반대로 180도의 각도는 코사인 유사도가 -1이 되며 90도는 0이 된다

1은 서로 유사도가 매우 높음을 의미하고, 0은 관련성 없음, -1은 완전히 반대됨을 의미한다



두 변수의 벡터 내적을 구한다

그 다음 두 벡터의 각각의 크기ㅇ를 구하여 곱한 값으로 나눈다

분자 부분은 벡터 내적으로, 각 변수들을 순서대로 곱한 다음 그 결과들을 모두 더한다

분모 부분은 벡터의 크기(norm)을 곱한 것으로 피타고라스의 정리에서 직각삼각형의 빗변을 구하는 것과 같다

코사인 유사도는 크기, 무게와 같은 변수간의 크기가 중요한 경우에는 적합하지 않은 거리 측정 방법이다

변수 간 스케일이 아닌 각 관측치의 스케일이 다를 때 코사인 유사도가 좋은 성능을 나타낸다

그리고 검색엔진의 경우, 질의어로 각 문서의 유사한 것을 찾는데 문서마다 길이 차이가 크기 때문에 코사인 유사도가 적합하다

코사인 유사도를 코사인 거리로 환산하는 방법은 간단하다 1에서 코사인 유사도를 빼주면 된다

유사도가 높을 수록 거리는 줄어들게 된다

# 데이터 거리 측정 실습

# 1. 패키지 임포트

```python
from scipy.spatial import distance
import pandas as pd
```

# 1. 데이터 불러오기 및 확인

```python
df= pd.read_csv('/content/Starbucks_Seoul.csv')

df.head()
```

# 3. 금천구에 위치한 스타벅스 매장 추출


```python
df1 = df[(df['gu_name'])=='금천구']

df1
```

# 4. 유클리드 거리 측정

```python
# 유클리드 거리
dist_euclidean = distance.cdist(df1[['latitude', 'longitude']],
                                df1[['latitude', 'longitude']], metric='euclidean')

# 칼럼명, 인덱스 설정
dist_euclidean = pd.DataFrame(data=dist_euclidean, columns=df1['name'])
dist_euclidean.set_index(df1['name'], inplace=True)

dist_euclidean
```

# 5. 맨해튼 거리 측정

```python
dist_manhattan = distance.cdist(df1[['latitude', 'longitude']],
                                df1[['latitude', 'longitude']], metric='cityblock')

# 칼럼명, 인덱스 설정
dist_manhattan = pd.DataFrame(data=dist_manhattan, columns=df1['name'])
dist_manhattan.set_index(df1['name'], inplace=True)

dist_manhattan
```

# 6. 민코프스키 거리 측정

```python
dist_minkowski = distance.cdist(df1[['latitude', 'longitude']],
                                df1[['latitude', 'longitude']], metric='minkowski', p=1.5)

# 칼럼명, 인덱스 설정
dist_minkowski = pd.DataFrame(data=dist_minkowski, columns=df1['name'])
dist_minkowski.set_index(df1['name'], inplace=True)

dist_minkowski
```

민코프스키 거리는 p값을 1로 설정하면 맨해튼 거리와 동일하고, 2로 설정하면 유클리드 거리와 동일하
다

# 7. 체비쇼프 거리 측정

```python
dist_chebyshev = distance.cdist(df1[['latitude', 'longitude']],
                                df1[['latitude', 'longitude']], metric='chebyshev')

# 칼럼명, 인덱스 설정
dist_chebyshev = pd.DataFrame(data=dist_chebyshev, columns=df1['name'])
dist_chebyshev.set_index(df1['name'], inplace=True)

dist_chebyshev
```

# 8. 마할라노비스 거리 측정

```python
dist_mahalanobis = distance.cdist(df1[['latitude', 'longitude']],
                                df1[['latitude', 'longitude']], metric='mahalanobis')

# 칼럼명, 인덱스 설정
dist_mahalanobis = pd.DataFrame(data=dist_mahalanobis, columns=df1['name'])
dist_mahalanobis.set_index(df1['name'], inplace=True)

dist_mahalanobis
```
