# 1. 패키지 임포트

```python
import scipy.stats
from scipy.stats import chi2_contingency
import pandas as pd
import matplotlib.pyplot as plt
```

# 2. 데이터 불러오기 및 확인

```python
df = pd.read_csv('/content/smoker.csv')

df.head()
```

```python
df.groupby(['sex','smoke'])['smoke'].count()
```

# 3. 카이제곱 검정용 데이터셋 가공

```python
crosstab = pd.crosstab(df.sex, df.smoke)
crosstab
```

# 4. 성별  흡연자 수 시각화

```python
%matplotlib inline
crosstab.plot(kind='bar', figsize=[10,5])
plt.grid()
```

# 5. 카이제곱 검정 수행

```python
chiresult = chi2_contingency(crosstab, correction = False)
print(f'Chi square: {chiresult[0]}')
print(f'P-value: {chiresult[1]}')
```
