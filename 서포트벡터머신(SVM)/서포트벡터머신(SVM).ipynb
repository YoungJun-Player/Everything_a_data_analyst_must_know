# 1. 필요한 패키지 임포트



```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import sklearn.svm as svm
from sklearn.svm import SVC
import pandas as pd
```

# 2. 데이터 불러오기 및 확인

```python
df = pd.read_excel('/content/drive/MyDrive/데이터분석/데이터 반드시 알아야 할 모든 것/6. 머신러닝/서포트벡터머신(SVM)/Raisin_Dataset.xlsx')

df.head()
```

# 3. 독립변수, 종속변수 테이블 분리 및 데이터 정규화

```python
# 독립변수, 종속변수 분리
df_x = df.drop(['Area','Class'], axis =1)
df_y = df[['Class']]

# 데이터 정규화 적용
MinMaxScaler = MinMaxScaler()
df_minmax = MinMaxScaler.fit_transform(df_x)

# 칼럼명 결합
df_x = pd.DataFrame(data=df_minmax, columns=df_x.columns)

df_x.head()
```

SVM은 거리를 기반으로 한 모델이기 때문에 스케일링을 필수적으로 해야 한다

# 4. 학습셋, 데이터셋 분리

```python
# 6:4로 생성
x_train, x_test, y_train, y_test = train_test_split(
    df_x, df_y, test_size = 0.4, random_state= 10)

# 잘 나뉘었는지 확인
print('train data 개수: ', len(x_train))
print('test data 개수: ', len(x_test))
```

# 5. 선형 svm 모델 성능 확인

```python
# 3차항 (degree), 학습 반복 횟수(max_iter) 10000 설정
svm_model = SVC(kernel = 'linear', degree= 3, gamma = 'auto', C=10, max_iter = 1000)    # max_iter 수가 너무 적으면 과소적합이 될 수 있다

svm_model.fit(x_train, y_train)

print(svm_model.score(x_train, y_train))
print(svm_model.score(x_test, y_test))
```

# 6. rbf SVM C값 1 ~30 별 모델 성능 확인

```python
scores =[]

for C_point in [*range(1,31)]:
    svc = SVC(kernel='rbf', C=C_point,max_iter = 1000)
    C_model = svc.fit(x_train, y_train)
    train_score = C_model.score(x_train, y_train)
    test_score = C_model.score(x_test, y_test)
    print(f'rbf SVM : c: {C_point}, train set score : {train_score}, test set score : {test_score}')

    scores.append([train_score, test_score])
```

c 값이 10에서 절정을 찍었다

# 7. rbf SVM gamma값 0.1~20별 모델 성능 확인

```python
scores =[]

for gamma_point in [*range(1,31)]:
    svc = SVC(kernel='rbf', C=10, gamma = gamma_point,max_iter = 1000)
    model = svc.fit(x_train, y_train)
    train_score = model.score(x_train, y_train)
    test_score = model.score(x_test, y_test)
    print(f'rbf SVM : gamma: {gamma_point}, train set score : {train_score}, test set score : {test_score}')

    scores.append([train_score, test_score])
```

5 ~10 정도에서 높은 정확성을 보인다
