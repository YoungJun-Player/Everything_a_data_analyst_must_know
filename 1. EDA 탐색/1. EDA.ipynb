7. 시각화를 통해 특정 변수의 분포 확인

```python
sns.displot(df['lead_time'])
#lead_time은 예약날짜로부터 투숙 날짜까지의 일수 차이를 의미
```

0 값이 확연히 많은 것으로 보아 당일체크 투숙객이 많은 것을 확인할 수 있다

하지만 실제로 당일 예약이라 그런 것인지 시스템상 기록이 제대로 남지 않을 것을 일괄적으로 0으로 입력한 것인지 확인할 필요가 있다.

8. 그룹 구분에 따른 특정 변수 분포 차이 시각화

```python
sns.violinplot(x='hotel', y='lead_time', data = df, inner=None, color= '.8')
sns.stripplot(x='hotel', y='lead_time', data = df, size = 1)
```

violinplot()함수는 분포를 효과적으로 표현해주며

stripplot()함수는 각 관측치의 위치를 직관적으로 표현해준다

resort호텔은 리드타임 최댓값이 city호텔보다 높지만 대체로 더 작은 값에 분포하는 것을 알 수 있다
