독립 변수들이 서로 단위가 다르거나 편차가 심할때

값의 스케일을 일정한 수준으로 변환시켜주는 표준화와 정규화 스케일링을 한다

독립변수 간 단위가 다른 것은 회귀분석이나 의사결정 나무 등 대부분의 데이터 마이닝 모델의 분석 결과에 별다른 영향을 미치지 않기 때문에

그대로 사용해도 무방하다

그러나 표준화나 정규화는 특정 머신러닝 모델의 학습 효율을 증가시키기 때문에 많이 사용된다

# 표준화


표준화는 각 관측치의 값이 전체 평균을 기준으로 어느 정도 떨어져 있는지 나타낼 때 사용한다

평균은 0으로 변환되고 1표준편차 거리는 +-1, 2표준편차거리는 +-2로 변환된다

zero-mean으로부터 얼마나 떨어져 있는지를 나타내기 때문에 이를 Z-score라 표현한다

표준화는 각 관측치 값에서 평균을 뺴준 후 표준편차로 나눠주면 된다



# 정규화

정규화는 데이터의 범위를 0부터 1까지로 변환하여 데이터 분포를 조정하는 방법이다

전체 데이터 중에서 해당 값이 어느 위치에 있는지 파악하는 데에 유용하다

0에 가까울수록 작은 값이고 1에 가까울 수록 큰 값이다

정규화는 (해당 값 - 최솟값) / (최댓값 - 최솟값)을 해주면 된다

표준화나 정규화는 k- Nearest Neighbor, 서포트벡터머신(SVM)과같은거리를활용한군집분석에서 필수적이다.

마찬가지로 범주화 알고리즘 인공신경망(Neural network) 모델에서도 학습 효율과 분류 성능을 높이기 위해 표준화나 정규화를 필수적으로 수행해줘야한다.

표준화와 정규화는 비슷한 듯하지만 다른 특성을 가지고 있기 때문에 목적에 따라 선택해야한다

이 외에 기본 표준화, 정규화 방식은 이상치에 민감하다는 단점을 보완한 스케일링 기법인 RotbustScaler도 많이 쓰인다

RobustScaler는 데이터의 중앙값(Q2)을 0으로 잡고 Q1(25%)과 Q3(75%) 사문위수와의 IQR차이를 1이 되도록 하는 스케일링 기법이다

이상치의 영향력을 최소화하여 일반적으로 표준화, 정규화보다 성능이 우수한 것으로 알려져있다

# 데이터 표준화와 정규화 스케일링 실습

# 1. 패키지 임포트

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# 2. 데이터 불러오기 및 확인

```python
df = pd.read_csv('/content/wine-clustering.csv')

df.head()
```

# 3. 각 칼럼의 평균과 분산 확인

```python
# 각 칼럼의 평균값
print(df.mean())
print('/n')

# 각 칼럼의 분산값
print(df.var())
```

# 4. 전체 칼럼 표준화 스케일링 적용

```python
# 데이터 표준화 적용
StandardScaler = StandardScaler()
df_stand = StandardScaler.fit_transform(df)

# 칼럼명 결함
df_stand = pd.DataFrame(data=df_stand, columns = df.columns)

df_stand.head()
```

StandardScaler() 함수를 사용하여 데이터셋에 적용하면 표준화된 값으로 변환된다

변환된 값은 넘파이 ndarray형태로 저장되기 때문에 다시 판다스 데이터프레임으로 변환해준다

# 5. 칼럼별 표준화 값의 평균과 분산 확인

```python
# 각 칼럼의 평균값
print(df_stand.mean())
print('\n')

# 각 칼럼의 분산값
print(df_stand.var())
```

# 7. 마그네슘 칼럼 스케일링 전과 후 분포 시각화

```python
%matplotlib inline
sns.displot(df['Magnesium'], height= 5, aspect=3)
sns.displot(df_stand['Magnesium'], height = 5, aspect=3)

plt.show()
```

표준화 전 분포 형태는 거의 유사하지만 데이터 값의 범위가 달라졌다

# 8. 전체 칼럼 정규화 스케일링 적용

```python
MinMaxScaler = MinMaxScaler()
df_minmax = MinMaxScaler.fit_transform(df)

# 칼럼명 결합
df_minmax = pd.DataFrame(data=df_minmax, columns = df.columns)

df_minmax.head()
```

# 9. 칼럼별 정규화 값의 최솟값과 최댓값 확인

```python
print(df_minmax.min())
print('\n')
print(df_minmax.max())
```

정규화는 모든 데이터 값의 범위를 0과 1 사이로 변환하기 때문에 최솟값은 0, 최댓값은 1로 변하였다

# 10. 정규화 값의 평균과 분산 확인

```python
print(df_minmax.mean())
print('\n')
print(df_minmax.var())
```

```python
# 11. 정규화 스케일링 전과 후 분포 시각화
%matplotlib inline
sns.displot(df['Magnesium'], height=5, aspect=3)
sns.displot(df_minmax['Magnesium'], height=5, aspect=3)

plt.show()
```

# 11. 전체 칼럼 RobustScaler 스케일링 적용

```python
# 데이터 RobustScaler 스케일링 적용
RobustScaler = RobustScaler()
df_robust = RobustScaler.fit_transform(df)

# 칼럼명 결함
df_robust = pd.DataFrame(data=df_robust, columns = df.columns)

df_robust.head()
```

# 12. RobustScaler 값의 평균과 분산 확인

```python
# 각 칼럼의 평균값
print(df_robust.mean())
print('\n')

# 각 칼럼의 분산값
print(df_robust.var())
```

RobustScaler는 평균 대신 중앙값을 사용하기 때문에 극단값의 영향을 거의 받지 않게 된다

# 13. RobustScaler 적용 전과 후 분포 비교


```python
%matplotlib inline
sns.displot(df['Magnesium'], height = 5, aspect = 3)
sns.displot(df_robust['Magnesium'], height = 5, aspect = 3)

plt.show()
```

이상치의 영향을 줄였기 때문에 전체 범위가 감소하였다
